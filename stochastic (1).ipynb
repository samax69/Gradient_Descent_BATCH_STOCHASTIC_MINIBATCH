```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_diabetes
```

```python
X , y = load_diabetes(return_X_y = True)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train , X_test , y_train , y_test = train_test_split(X , y , test_size = 0.2)
```

```python
class SGD:
    def __init__(self , learning_rate , epochs):
        self.lr = learning_rate
        self.epochs = epochs
        self.coef_ = None
        self.intercept_ = None

    def fit(self , X_train , y_train):
        self.coef_ = np.ones(X_train.shape[1])
        self.intercept_ = 0
        for i in range(self.epochs):
            for 
            idx = np.random.randint(0 , X_train.shape[0])
            y_hat = np.dot(X_train[idx] , self.coef_) + self.intercept_
            intercept_derv = -2 * (y_train[idx] - y_hat)
            self.intercept_ = self.intercept_ - (self.lr * intercept_derv)

            coef_derv = -2 * (y_train[idx] - y_hat) * X_train[idx]
            self.coef_  = self.coef_ - (self.lr * coef_derv)
        print(self.coef_)
        print(self.intercept_)
    def predict(self , X_test):
        return np.dot(X_test , self.coef_) + self.intercept_
            
```

```python
reg = SGD(learning_rate = 0.1 , epochs = 1000)
```

```python
reg.fit(X_train , y_train)
```

```python
y_pred = reg.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error , r2_score
```

```python
mean_squared_error(y_test, y_pred)
r2_score(y_test , y_pred)
```
